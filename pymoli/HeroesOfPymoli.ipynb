```python
import pandas as pd

# read adat from CSv
purchase_data = pd.read_csv("../resources/purchase_data.csv")
purchase_data.head()
```

# Player Count

```python
total_players = len(purchase_data['SN'].unique())
summary_table1 = pd.DataFrame({'Total Players': [total_players]})

summary_table1
```

# Purchasing Analysis (Total)

```python
# Calculate puchasing stats
unique_items = len(purchase_data['Item ID'].unique())
average_price = round(purchase_data['Price'].mean(), 2)
num_purchases = len(purchase_data['Purchase ID'].unique())
total_rev = purchase_data['Price'].sum()

# Create table
summary_table2 = pd.DataFrame({
    'Number of Unique Items': [unique_items], 
    'Average Price': ["$" + str(average_price)],
    'Number of Purchases': [num_purchases],
    'Total Revenue': ["$" + str(total_rev)]
})

summary_table2
```

# Gender Demographics

```python
# Calculate demographic stats
demo_data = purchase_data[['SN', 'Age', 'Gender']]
demo_data = demo_data.drop_duplicates()
gd_data = demo_data.groupby(['Gender'])

gd_count = gd_data['SN'].count()
gd_per = gd_count / total_players * 100

# Create table
summary_table3 = pd.DataFrame({
    'Total Count': gd_count,
    'Percentage of Players': gd_per
})

summary_table3['Percentage of Players'] = summary_table3['Percentage of Players'].map('{:,.2f}%'.format)

summary_table3
```

# Purchasing Analysis (Gender)

```python
# Group data by gender and calculate purchasing stats
gp_data = purchase_data.groupby(['Gender'])

gp_count = gp_data['SN'].count()
gp_price = gp_data['Price'].mean()
gp_total = gp_data['Price'].sum()

# Create table
summary_table4 = pd.DataFrame({
    'Purchase Count': gp_count,
    'Average Purchase Price': gp_price,
    'Total Purchase Value': gp_total,
    'Average Total Purchase per Person': gp_total / summary_table3['Total Count']
})

# Reformat table
summary_table4['Average Purchase Price'] = summary_table4['Average Purchase Price'].map('${:,.2f}'.format)
summary_table4['Total Purchase Value'] = summary_table4['Total Purchase Value'].map('${:,.2f}'.format)
summary_table4['Average Total Purchase per Person'] = summary_table4['Average Total Purchase per Person'].map('${:,.2f}'.format)
               
summary_table4
```

# Age Demographics

```python
ad_data = demo_data

# Group data into bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
ad_data['Age Group'] = pd.cut(ad_data['Age'], bins, labels=age_groups)

# Group by bins and fins demographic stats
ad_data = ad_data.groupby(['Age Group'])

ad_count = ad_data['SN'].count()
ad_per = ad_count / total_players * 100

# Table
summary_table5 = pd.DataFrame({
    'Total Count': ad_count,
    'Percentage of Players': ad_per
})

# Format
summary_table5['Percentage of Players'] = summary_table5['Percentage of Players'].map('{:,.2f}%'.format)

summary_table5
```

# Purchasing Analysis (Age)

```python
ap_data = purchase_data

# Use bins to group data
ap_data['Age Group'] = pd.cut(ap_data['Age'], bins, labels=age_groups)

ap_data = ap_data.groupby(['Age Group'])

ap_count = ap_data['SN'].count()
ap_price = ap_data['Price'].mean()
ap_total = ap_data['Price'].sum()

# Table and formatting
summary_table6 = pd.DataFrame({
    'Purchase Count': ap_count,
    'Average Purchase Price': ap_price,
    'Total Purchase Value': ap_total,
    'Average Total Purchase per Person': ap_total / summary_table5['Total Count']
})

summary_table6['Average Purchase Price'] = summary_table6['Average Purchase Price'].map('${:,.2f}'.format)
summary_table6['Total Purchase Value'] = summary_table6['Total Purchase Value'].map('${:,.2f}'.format)
summary_table6['Average Total Purchase per Person'] = summary_table6['Average Total Purchase per Person'].map('${:,.2f}'.format)
               
summary_table6
```

# Top Spenders

```python
# Group by user and calculate top spenders
s_data = purchase_data.groupby(['SN'])

sp_count = s_data['SN'].count()
sp_price = round(s_data['Price'].mean(), 2)
sp_total = s_data['Price'].sum()

# Create table
summary_table7 = pd.DataFrame({
    'Purchase Count': sp_count,
    'Average Purchase Price': sp_price,
    'Total Purchase Value': sp_total,
})

# Restrict to top five rows and format
summary_table7 = summary_table7.sort_values('Total Purchase Value', ascending=False).head(5)

summary_table7['Average Purchase Price'] = summary_table7['Average Purchase Price'].map('${:,.2f}'.format)
summary_table7['Total Purchase Value'] = summary_table7['Total Purchase Value'].map('${:,.2f}'.format)

summary_table7
```

# Most Popular Items

```python
# Group by item and determine poularity stats
i_data = purchase_data.groupby(['Item ID', 'Item Name'])

i_count = i_data['SN'].count()
i_price = round(i_data['Price'].mean(), 2)
i_total = i_data['Price'].sum()

# Create table
summary_table8 = pd.DataFrame({
    'Purchase Count': i_count,
    'Item Price': i_price,
    'Total Purchase Value': i_total,
})

# Sort descending and format
summary_table9 = summary_table8.sort_values('Purchase Count', ascending=False).head(5)

summary_table9['Item Price'] = summary_table9['Item Price'].map('${:,.2f}'.format)
summary_table9['Total Purchase Value'] = summary_table9['Total Purchase Value'].map('${:,.2f}'.format)

summary_table9
```

# Most Profitable Items

```python
# Sort to create table of most proftiable items 
summary_table10 = summary_table8.sort_values('Total Purchase Value', ascending=False).head(5)

# Format
summary_table10['Item Price'] = summary_table10['Item Price'].map('${:,.2f}'.format)
summary_table10['Total Purchase Value'] = summary_table10['Total Purchase Value'].map('${:,.2f}'.format)

summary_table10
```
